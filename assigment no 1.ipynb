```python
salary = float(input("Enter your salary: "))
years = float(input("Enter your years of service: "))

if years > 5:
    bonus = salary * 0.05
    print("Your net bonus amount is:", bonus)
else:
    print("No bonus. Service is less than or equal to 5 years.")
```

```python
marks = int(input("Enter your marks: "))

if marks < 25:
    print("Grade: F")
elif marks < 45:
    print("Grade: E")
elif marks < 50:
    print("Grade: D")
elif marks < 60:
    print("Grade: C")
elif marks <= 80:
    print("Grade: B")
else:
    print("Grade: A")
```
